```python
import sys
#!{sys.executable} -m pip install requests
#!{sys.executable} -m pip install html5lib
#!{sys.executable} -m pip install bs4
#!{sys.executable} -m pip install bs4 selenium
import re
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36',
'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
'Accept-Language': 'en-US,en;q=0.5',
'Accept-Encoding': 'gzip, deflate',
'Connection': 'keep-alive',
'Upgrade-Insecure-Requests': '1',
}
s = requests.Session()
s.headers.update(headers)

comptage_produit = 0
nom_produit=[]
prix_produit=[]
note_produit=[]
nb_avis_produit=[]
commentaire =[]
note_comm=[]
date_com=[] 
marques=[]


url_base_pdt = 'https://www.dafy-moto.com'
url_base_page = 'https://www.dafy-moto.com/casques.html?brand=5,6,82,98,144,148,151&p='
url_page = "https://www.dafy-moto.com/casques.html?brand=5,6,82,144,148,151,98"

url_pages=[]
url_pages.append('https://www.dafy-moto.com/casques.html?brand=5,6,82,144,148,151,98')
for i in range(2,50):
    url_page = url_base_page + str(i)
    url_pages.append(url_page)

for i in range(0,49):
    url_page_2 = url_pages[i] 
    print(url_page_2)
    
    r = s.get(url_page_2,allow_redirects= False)
    time.sleep(1)
    soup = BeautifulSoup(r.text, 'html.parser')
    produits = soup.find('div',attrs={'id':"es-products"})
    produits = produits.find('div',attrs={'id':"product-results"})
    produits = produits.find_all('div',attrs={'class':"block-product__item js-block-product-item"})

    for produit in produits :
        comptage_produit +=1
        print(comptage_produit)
        try : 
            url_produit_2 = produit.find('a')['href']
            url_produit = url_base_pdt + url_produit_2
        except : 
            pass 
        try :
            marque = produit.find('div',attrs={'class':"product-single__brand"}).text
        except :
            marque = ""
        
        print(url_produit)
        print(marque)
        browser = webdriver.Chrome(executable_path = r"C:\Users\basti\OneDrive\Documents\1. Etudes\4.Stages\1. Eleven\2. Projet\2. Scrapping et SA\chromedriver.exe")
        user = ""
        pwd = ""
        browser.get(url_produit)
        time.sleep(1)
        soup = BeautifulSoup(browser.page_source, 'html.parser')
        p=1
        try : 
            nb_avis = soup.find('meta',attrs={'itemprop':"reviewCount"})['content']
        except : 
            nb_avis = '0'
        
        nb_page = float(nb_avis) - 1
        nb_page = (float(nb_page)/3) 
        nb_pages = int(nb_page) + 1

        while p <= nb_pages : 
            
            try :
                element = browser.find_element_by_xpath("//span[@class='page-link js-page-link'][@data-page="+ str(p)+"]") 
                time.sleep(1)
                browser.execute_script("arguments[0].click();", element)
            except : 
                pass
            soup = BeautifulSoup(browser.page_source, 'html.parser')

            p += 1
            
            try : 
                nom = str(soup.find('h1',attrs={'class':"product-article__title"}).text)
                nom = re.sub(r"[^0-9a-zA-Z]"," ",nom)
            except :
                nom="-"
            try :
                prix = soup.find('div',attrs={'class':"product-price-line"}).text
            except : 
                prix="-"
            
            try : 
                note_pdt=soup.find('meta',attrs={'itemprop':"ratingValue"})['content']
            except : 
                note_pdt = "-"
            
            try :
                avis = soup.find('div',attrs={'class':"avis-verifies-wrapper"})
                avis = avis.find_all('div',attrs={'class':"avis-verifies"})
            except AttributeError :
                pass
            try : 
                for avi in avis :
                    marques.append(marque)
                    nb_avis_produit.append(nb_avis)
                    nom_produit.append(nom)
                    note_produit.append(note_pdt)
                    prix_produit.append(prix)
                    try : 
                        com = avi.find('div',attrs={'class':"avis-verifies__comment"}).text
                    except :
                        com ="-"
                    try :
                        date = avi.find('span',attrs={'class':"rating__date"}).text
                    except : 
                        date ="-"
                    commentaire.append(com)
                    date_com.append(date)
                    print(com)
                    print(date)
                    try :
                        rating = 0
                        notes = avi.find('div',attrs={'class':"rating"})
                        notes = avi.find_all('i')
                        for note in notes : 
                            if note['class']==['icon-dafy-star', 'rating-star', 'first-star']:
                                rating += 1
                            elif note['class']==['icon-dafy-star', 'rating-star']:
                                rating += 1
                    except : 
                        rating = '-'
                    print(rating)
                    note_comm.append(rating)
            except TypeError :
                pass
        browser.quit()
        print(p)
    print(nb_avis)
```

```python
from pandas import DataFrame

Avis_Dafy = {'Produit': nom_produit,
                  'Prix': prix_produit,
                  'Note': note_produit,
                  'Nombre avis': nb_avis_produit,
                  'Commentaires': commentaire,
                  'Note du commentaire': note_comm,
                  'Date': date_com,
                  'Marque': marques
                  }

df = DataFrame(Avis_Dafy, columns= ['Produit', 'Prix','Note','Nombre avis','Commentaires','Note du commentaire','Date','Marque'])
```
