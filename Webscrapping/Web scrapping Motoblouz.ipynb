```python
import sys
#!{sys.executable} -m pip install requests
#!{sys.executable} -m pip install html5lib
#!{sys.executable} -m pip install bs4
#!{sys.executable} -m pip install bs4 selenium
import re
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python

comptage_produit=[]
nom_produit=[]
prix_produit=[]
note_produit=[]
nb_avis_produit=[]
commentaire =[]
note_comm=[]
avis_g=[]
date_com=[] 
links=[]
marques=[]



url_page_base = "https://www.motoblouz.com/achat-casque-5695/marque/agv/marque/airoh/marque/hjc/marque/ls2/marque/nolan/marque/scorpion-exo/marque/shark/marque/shoei/"
url_pages=[]
url_pages.append('https://www.motoblouz.com/achat-casque-5695/marque/agv/marque/airoh/marque/hjc/marque/ls2/marque/nolan/marque/scorpion-exo/marque/shark/marque/shoei.html')
for i in range(2,25):
    url_page = url_page_base +"page-"+ str(i) + ".html"
    url_pages.append(url_page)

for i in range(17,24):
    url_page_2 = url_pages[i] 
    print(url_page_2)


    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
    'Accept-Language': 'en-US,en;q=0.5',
    'Accept-Encoding': 'gzip, deflate',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1',
    }
    s = requests.Session()
    s.headers.update(headers)
    r = s.get(url_page_2,allow_redirects= False)
    url_base_2 = "https://www.motoblouz.com"
    soup = BeautifulSoup(r.text, 'html.parser')
    produits = soup.find('div',attrs={'class':"products"})
    produits = produits.find_all('div',attrs={'class':"product-inner"})
    
    for produit in produits :

        com_pdt=[]
        liste_produit = produit.find('a').attrs['href']
        url_base_pdt = url_base_2 + liste_produit
        marque = produit.find('div',attrs={'class':"brand-name"})
        marque = marque.text
        print('Marque:' + marque)
        
    
        s = requests.Session()
        s.headers.update(headers)
        r = s.get(url_base_pdt,allow_redirects= False)
    
        time.sleep(0.5)
        soup = BeautifulSoup(r.text, 'html.parser')

    
    
        # extraction nom du produit
        try :
            desc = soup.find('div', attrs={'class':"block has-border title-and-desc"})
            nom = desc.find('h1')
            nom = nom.text
        except AttributeError:
            nom = '-'
        print('Nom du produit:'+ nom)
    
        # extraction prix 
        try :
            prix = soup.find('span',attrs={'class':"price"})['content']
        except AttributeError:
            prix = '0'
        except TypeError:
            prix='0'
        print('Prix:' + prix)
    
        #extraction note
        try :
            note = soup.find('span',attrs={'class':"hidden"})
            note = note.text
        except AttributeError:
            note = '-'
        print(note)
    
        #extraction nombre avis
        try :
            nb_avis = soup.find('a',attrs={'class':"d3t-smooth-anchor"})
            nb_avis = nb_avis.text
            nb_avis = nb_avis.replace('avis','')
            nb_avis = nb_avis.replace('(','')
            nb_avis = nb_avis.replace(')','')
            nb_avis = int(nb_avis)
        except AttributeError:
            nb_avis = '0'
        print(nb_avis)
    
        j = 0
    
        while int(len(com_pdt)) < int(nb_avis) :
            j = j + 1
            url = url_base_pdt  + '?bvstate=pg:'+ str(j)+'/ct:r#'
            print(url)
            
            
            browser = webdriver.Chrome(executable_path = r"C:\Users\Bastien\eleven\Engagements - 2RH\5. Analyses\3. Data - Bastien\1. Webscrapping\chromedriver.exe")
            user = ""
            pwd = ""
            browser.get(url)
            time.sleep(1)
            soup = BeautifulSoup(browser.page_source, 'html.parser')
            browser.quit()


        
            # on trouve les avis
            review = soup.find('div',attrs={'id':'reviews'})
            review_d1 = review.find_all('div',attrs={'class':"bv-content-container"})
            
            if len(review_d1)==0 :
                nb_avis = '0'
            elif len(review_d1) != 0 :

            
                for comment in review_d1:
                    nom_produit.append(nom)
                    prix_produit.append(prix)
                    note_produit.append(note)
                    nb_avis_produit.append(nb_avis)
                    marques.append(marque)
                    
                
                
                    try :
                        avis=comment.find('div',attrs={'class':"bv-content-summary-body-text"})
                        avis=avis.text
                        commentaire.append(avis)
                        com_pdt.append(avis)
                    except AttributeError : 
                        commentaire.append('')
                        com_pdt.append('')
                
                    try :
                        note_com_ws = comment.find('meta',attrs={'itemprop':"ratingValue"})
                        note_com_ws = note_com_ws['content']
                        note_comm.append(note_com_ws)
                    except TypeError:
                        note_comm.append('')
            
                    try :
                        avis_g_ws = comment.find('div',attrs={'class':"bv-content-title-container"})
                        avis_g_ws = avis_g_ws.find('h4')
                        avis_g_ws = avis_g_ws.text
                        avis_g.append(avis_g_ws)
                    except AttributeError:
                        avis_g.append('')    
            
                    try : 
                        date_com_ws = comment.find('meta',attrs={'itemprop':"datePublished"})
                        date_com_ws = date_com_ws['content']
                        date_com.append(date_com_ws)
                    except TypeError:
                        date_com.append('')
```

```python
from pandas import DataFrame

Avis_Motoblouz = {'Produit': nom_produit,
                  'Prix': prix_produit,
                  'Note': note_produit,
                  'Nombre avis': nb_avis_produit,
                  'Commentaires': commentaire,
                  'Note du commentaire': note_comm,
                  'Avis Général': avis_g,
                  'Date': date_com,
                  'Marque': marques
                  }

df = DataFrame(Avis_Motoblouz, columns= ['Produit', 'Prix','Note','Nombre avis','Commentaires','Note du commentaire','Avis Général','Date','Marque'])
```
