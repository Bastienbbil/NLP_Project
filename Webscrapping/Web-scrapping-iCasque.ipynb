```python
import sys
#!{sys.executable} -m pip install requests
#!{sys.executable} -m pip install html5lib
#!{sys.executable} -m pip install bs4
#!{sys.executable} -m pip install bs4 selenium
import re
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
list_url_pdt =[]
url = 'https://www.icasque.com/Casque-moto/AGV/?m=22,10,19,238,11,34,51'
browser = webdriver.Chrome(executable_path = r"C:\Users\basti\OneDrive\Documents\1. Etudes\4.Stages\1. Eleven\2. Projet\2. Scrapping et SA\chromedriver.exe")
user = ""
pwd = ""
browser.get(url)
time.sleep(5)
p=1

for i in range(1,50):
    print(p)
    p +=1
    time.sleep(2)
    
    soup = BeautifulSoup(browser.page_source, 'html.parser')
    produits = soup.find('div',attrs={'class':"products row"})
    produits = produits.find_all('article',attrs={'class':"col-prod productList"})
    for produit in produits :
        url_pdt = produit.find('a',attrs={'itemprop':"url"})['href']
        print(url_pdt)
        list_url_pdt.append(url_pdt)
    element = browser.find_element_by_xpath("//div[@class='flex flex-ma-between flex-ca-center flex-b100 pt-15 pb-40 pagination-box']/a[2]") 
    time.sleep(1)
    browser.execute_script("arguments[0].click();", element)
    time.sleep(1)

browser.quit()
```

```python
comptage_produit=[]
nom_produit=[]
prix_produit=[]
note_produit=[]
nb_avis_produit=[]
commentaire =[]
note_comm=[]
avis_g=[]
date_com=[] 
links=[]
marques=[]

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36',
'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
'Accept-Language': 'en-US,en;q=0.5',
'Accept-Encoding': 'gzip, deflate',
'Connection': 'keep-alive',
'Upgrade-Insecure-Requests': '1',
}
s = requests.Session()
s.headers.update(headers)

for url in list_url_pdt:
    
    print(url)
    r = s.get(url,allow_redirects= False)
    time.sleep(2)
    soup = BeautifulSoup(r.text, 'html.parser')
    try : 
        marque_1 = soup.find('section', attrs={'class':"breadcrumb container d-none d-lg-block"})
        marque = marque_1.find_all('span', attrs={'itemprop':"name"})[3].text
    except : 
        marque = 'Marque non disponible'
    print(marque)

    try : 
        nom = marque_1.find_all('span', attrs={'itemprop':"name"})[4].text
    except : 
        nom = 'Nom indisponible'
    print(nom)

    try :
        note = soup.find('div',attrs={'class':"prod-stars fs11 rating rating-lg rating45"})
        note = note.find('meta',attrs={'itemprop':"ratingValue"})['content']
    except AttributeError : 
        note='0'
    print(note)
            # extraction prix 
    try :
        prix = soup.find('span',attrs={'class':"price"})['content']
    except : 
        prix = 'pas de prix'
    print(prix)

    try :
        nb_avis = soup.find('span',attrs={'itemprop':"reviewCount"})
        nb_avis = nb_avis.text
        nb_avis = int(nb_avis)
    except AttributeError : 
        nb_avis = 0
    print(nb_avis)


    # on trouve les avis
    try :
        review = soup.find('div',attrs={'id':"reviews-list"})
        review_d1 = review.find_all('article',attrs={'itemprop':"review"})
    except AttributeError :
        review_d1 = []

    if len(review_d1)==0 :
        nb_avis = '0'
    elif len(review_d1) != 0 :

        for comment in review_d1:
            nb_avis_produit.append(nb_avis)
            note_produit.append(note)
            marques.append(marque)
            prix_produit.append(prix)
            nom_produit.append(nom)

            try : 
                avis=comment.find('span',attrs={'itemprop':"description"})
                avis=avis.text
                commentaire.append(avis)
                com_pdt.append(avis)

            except : 
                avis='-'
                commentaire.append(avis)
                com_pdt.append(avis)
            print(avis)    


            try : 
                note_com_ws = comment.find('div',attrs={'class':"box-rating global-rating"})
                note_com_ws = comment.find('span',attrs={'class':"rating-value"})
                note_com_ws = note_com_ws.text
                note_comm.append(note_com_ws)
            except : 
                note_com_ws = '0'
                note_comm.append(note_com_ws)
            print(note_com_ws)

            try :
                avis_g_ws = comment.find('span',attrs={'itemprop':"name"})
                avis_g_ws = avis_g_ws.text
                avis_g.append(avis_g_ws)
            except : 
                avis_g_ws='-'
                avis_g.append(avis_g_ws)
            print(avis_g_ws)

            try :
                date_com_ws = comment.find('meta',attrs={'itemprop':"datePublished"})
                date_com_ws = date_com_ws['content']
                date_com.append(date_com_ws)
            except : 
                date_com_ws = '-'
                date_com.append(date_com_ws)
            print(date_com_ws)
```

```python
from pandas import DataFrame

Avis_iCasque = {'Produit': nom_produit,
                  'Prix': prix_produit,
                  'Note': note_produit,
                  'Nombre avis': nb_avis_produit,
                  'Commentaires': commentaire,
                  'Note du commentaire': note_comm,
                  'Avis Général': avis_g,
                  'Date': date_com,
                  'Marque': marques
                  }

df = DataFrame(Avis_iCasque, columns= ['Produit', 'Prix','Note','Nombre avis','Commentaires','Note du commentaire','Avis Général','Date','Marque'])
```
